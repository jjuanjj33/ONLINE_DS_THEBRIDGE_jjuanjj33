```python
import pandas as pd
import geopandas as gpd

# Cargar archivo GeoJSON
gdf = gpd.read_file("C:/Users/34666/Documents/MASTER EN IA/THE BRIDGE/REPO_CLASE/REPO_PRACTICE/EDA/RESTAURANTS/data/export.geojson")

# Ver las primeras filas
print(gdf.head())

# Puedes convertirlo a DataFrame plano si no necesitas la geometría
df = pd.DataFrame(gdf.drop(columns='geometry'))
df
```

```python
# Umbral: 90% de valores nulos
umbral = 0.9

# Calcular porcentaje de nulos por columna
porcentaje_nulos = df.isnull().mean()

# Filtrar columnas que tienen menos o igual al 30% de nulos
columnas_buenas = porcentaje_nulos[porcentaje_nulos <= umbral].index

# Crear nuevo DataFrame con solo las columnas buenas
df_cleaned = df[columnas_buenas]
df_cleaned2 = df_cleaned.drop(["id", "@id", "addr:city", "amenity", "check_date", "phone", "website", "wheelchair", "opening_hours"], axis=1)
df_cleaned2
```

```python
# Combinar calle y número si existen
df_cleaned2["addr:street"] = df_cleaned2["addr:street"].fillna("")  # Reemplaza None por cadena vacía
df_cleaned2["addr:housenumber"] = df_cleaned2["addr:housenumber"].fillna("")

# Unir con coma si hay número
df_cleaned2["addr:street"] = df_cleaned2.apply(
    lambda row: f"{row['addr:street']}, {row['addr:housenumber']}".strip(", ") 
    if row["addr:housenumber"] else row["addr:street"],
    axis=1
)
df_cleaned2
df_cleaned3 = df_cleaned2.drop("addr:housenumber", axis=1)
df_cleaned3
```

```python

df_cleaned3 = df_cleaned3.rename(columns={"addr:street":"Direccion",
                                          "addr:postcode":"CP",
                                          "cuisine":"Tipo",
                                          "name":"Restaurante"
})
nuevo_orden = ["Restaurante", "Tipo", "Direccion", "CP"]
df_cleaned4 = df_cleaned3[nuevo_orden]
df_cleaned4
```
